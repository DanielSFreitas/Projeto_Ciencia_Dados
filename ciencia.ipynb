```python
df_clean_2['EDUCATION'].replace(to_replace=[0,5,6], value=4, inplace=True)
df_clean_2['EDUCATION'].value_counts()
```

```python
df_clean_2['MARRIAGE'].value_counts()
```

```python
df_clean_2['MARRIAGE'].replace(to_replace=0, value=3, inplace=True)
df_clean_2['MARRIAGE'].value_counts()
```

```python
df_clean_2['EDUCATION_CAT'] = 'none'
cat_maping = {
    1: "graduate school",
 2: "university",
 3: "high school",
 4: "others"

}
```

```python
df_clean_2['EDUCATION_CAT'] = df_clean_2['EDUCATION'].map(cat_maping)
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
edu_ohe = pd.get_dummies(df_clean_2['EDUCATION_CAT'], dtype=int)
edu_ohe.head(10)
```

```python
df_with_ohe = pd.concat([df_clean_2, edu_ohe], axis=1)
df_with_ohe[['EDUCATION_CAT', 'graduate school','high school','others','university']].head(10)
```

```python
df_with_ohe.to_csv('dataset_clean.csv', index=False)
```
